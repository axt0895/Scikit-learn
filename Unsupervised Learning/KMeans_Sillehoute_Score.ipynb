```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
X, y = make_blobs(centers=6, n_samples=700, n_features=2, random_state=42, shuffle=True)
```

```python
plt.scatter(X[:, 0], X[:, 1], s = 20)
```

```python
k_means = [KMeans(n_clusters=k, n_init='auto', random_state=42) for k in [3, 4, 5, 6, 7, 8]]
```

```python
for cluster in k_means:
    cluster.fit(X)
```

## Inertia Score for each cluster

```python
for i, cluster in enumerate(k_means):
    print(f'Cluster {i+1}', cluster.inertia_)
```

## Cluster Centriod:

```python
for i, cluster in enumerate(k_means):
    print(f'Cluster {i+1}', cluster.cluster_centers_)
```

## Sillehoutte Score

```python
scores = [silhouette_score(X, cluster.labels_) for cluster in k_means]
```

```python
plt.plot([3, 4, 5, 6, 7, 8], scores)
```

## Imports

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from mlxtend.plotting import plot_decision_regions
```

## Datasets

```python
X, y = make_blobs(n_samples=700,n_features=2, centers=5, random_state=42, shuffle=True)
```

```python
plt.scatter(X[:,0], X[:,1], s= 20)
plt.show()
```

```python
k_means = KMeans(n_clusters=5, n_init='auto', random_state=42)
k_means.fit(X)
```

## plotting decision regions for two different clusters

```python
k_means = [KMeans(n_clusters=4, n_init='auto', random_state=42), KMeans(n_clusters=5, n_init='auto', random_state=42)]
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
for kmean, axes in zip(k_means, ax):
    kmean.fit(X)
    plot_decision_regions(clf=kmean, X=X,y=y, ax=axes)
plt.show()
```

```python
import numpy as np
X_new = np.array([[0, 2],[3, 2],[-3, 3],[-5, 3.5]])
```

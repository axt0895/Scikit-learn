## Imports

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from mlxtend.plotting import plot_decision_regions
```

```python
iris = load_iris(as_frame=True)
```

```python
X, y = iris.data, iris.target
```

```python
decision_tree = DecisionTreeClassifier(max_depth=2, min_samples_leaf=20, random_state=42)
```

## Only taking petal length and petal width as input features

```python
X = X[['petal length (cm)', 'petal width (cm)']]
```

```python
decision_tree.fit(X, y)
```

```python
decision_tree
```

```python
decision_tree.feature_names_in_
```

```python
export_graphviz(
    decision_tree, 
    out_file='iris_tree.dot',
    feature_names=['petal length', 'petal width'],
    class_names=iris.target_names,
    rounded=True, 
    filled=True
)
```

```python
from graphviz import Source
```

```python
Source.from_file('iris_tree.dot')
```

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=700, noise=0.15, random_state=42)
```

```python
X.shape
```

```python
plt.scatter(X[:, 0], X[:,1], s=15, c=['red' if cat == 0 else 'blue' for cat in y])
```

```python
decision_tree_1 = DecisionTreeClassifier(random_state=42)
decision_tree_2 = DecisionTreeClassifier(max_depth=4, random_state=42)
decision_tree_3 = DecisionTreeClassifier(min_samples_leaf=5, random_state=42)
```

```python
decision_trees = [decision_tree_1, decision_tree_2, decision_tree_3]
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 5))
for i, tree in enumerate(decision_trees):
    tree.fit(X, y)
    plot_decision_regions(clf=tree, X=X, y=y, ax=ax[i])
    ax[i].set_title(f'Decision Tree {i}')
plt.show()
```
